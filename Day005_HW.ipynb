## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests, json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(r.text)

print('APIㄧ次回傳' + str(len(data)) + '筆資料')
print('每一筆資料包含以下欄位：')
for key, value in data[0].items():
    print(key) 
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for datum in data:
    print('標題：', datum['title'])
    print('貼文時間：', datum['createdAt'])
    print('留言人數：', datum['commentCount'])
    print('按讚人數：', datum['likeCount'])
    print("###################################")
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
comment_number = []
like_number = []
for datum in data:
    comment_number.append(datum['commentCount'])
    like_number.append(datum['likeCount'])
print('平均留言人數:',sum(comment_number)/len(comment_number))
print('平均按讚人數:',sum(like_number)/len(like_number))
```
