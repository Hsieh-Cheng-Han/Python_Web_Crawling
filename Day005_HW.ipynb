```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests, json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(r.text)

print('APIã„§æ¬¡å›å‚³' + str(len(data)) + 'ç­†è³‡æ–™')
print('æ¯ä¸€ç­†è³‡æ–™åŒ…å«ä»¥ä¸‹æ¬„ä½ï¼š')
for key, value in data[0].items():
    print(key) 
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for datum in data:
    print('æ¨™é¡Œï¼š', datum['title'])
    print('è²¼æ–‡æ™‚é–“ï¼š', datum['createdAt'])
    print('ç•™è¨€äººæ•¸ï¼š', datum['commentCount'])
    print('æŒ‰è®šäººæ•¸ï¼š', datum['likeCount'])
    print("###################################")
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
comment_number = []
like_number = []
for datum in data:
    comment_number.append(datum['commentCount'])
    like_number.append(datum['likeCount'])
print('å¹³å‡ç•™è¨€äººæ•¸:',sum(comment_number)/len(comment_number))
print('å¹³å‡æŒ‰è®šäººæ•¸:',sum(like_number)/len(like_number))
```
